```python
import pandas as pd
```

```python
df = pd.read_csv('./lettuce_price.csv')
df.drop(['Unnamed: 10'], axis=1, inplace=True)
df.head()
```

```python
df['產品'].unique()
```

```python
target2_col_name = 'LI7 萵苣菜 紅捲綠捲 '
target_df = df[df['產品'] == 'LI8 萵苣菜 水耕 ']
target2_df = df[df['產品'] == target2_col_name]
```

```python
target_df.loc[:,'日期'] = target_df.loc[:, '日期'].apply(lambda x: str(int(x.split('/')[0])+1911)+'/'+x.split('/')[1]+'/'+x.split('/')[2])
target2_df.loc[:,'日期'] = target2_df.loc[:, '日期'].apply(lambda x: str(int(x.split('/')[0])+1911)+'/'+x.split('/')[1]+'/'+x.split('/')[2])
```

```python
from matplotlib import pyplot as plt

target_df.loc[:, '日期'] = pd.to_datetime(target_df['日期'], format='%Y/%m/%d')
target_df.loc[:, '平均價(元/公斤)'] = target_df['平均價(元/公斤)'].astype('float')

plt.plot(target_df['日期'], target_df['平均價(元/公斤)'])
plt.title('Water lettuce price')
plt.show()
```

```python
target2_df.loc[:, '日期'] = pd.to_datetime(target2_df['日期'], format='%Y/%m/%d')
target2_df.loc[:, '平均價(元/公斤)'] = target2_df['平均價(元/公斤)'].astype('float')

plt.plot(target2_df['日期'], target2_df['平均價(元/公斤)'])
plt.title(target2_col_name)
plt.show()
```
